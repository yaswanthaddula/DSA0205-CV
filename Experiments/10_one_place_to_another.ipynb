```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/Screenshot 2026-02-26 135627.png")

if img is None:
    print("Image not loaded")
else:
    cv2_imshow(img)

    rows, cols = img.shape[:2]

    M1 = np.float32([[1, 0, 100],
                     [0, 1, 50]])

    moved1 = cv2.warpAffine(img, M1, (cols, rows))
    cv2_imshow(moved1)

    M2 = np.float32([[1, 0, -100],
                     [0, 1, -50]])

    moved2 = cv2.warpAffine(img, M2, (cols, rows))
    cv2_imshow(moved2)
```
