```python
#4 Dilate an Image using Dilate function.
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

image = cv2.imread('/content/Screenshot 2026-02-26 135502.png')
print("Original image")
cv2_imshow(image)

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Create kernel and apply dilation
kernel = np.ones((5,5), np.uint8)
dilated = cv2.dilate(gray_image, kernel, iterations=1)

# Display dilated image
cv2_imshow(dilated)
```
