```python
#Mean Filter
from google.colab.patches import cv2_imshow
import cv2
img = cv2.imread("/content/Screenshot 2026-02-26 140552.png")

# Apply Mean Filter
mean = cv2.blur(img, (3,3))
cv2_imshow(img)
cv2_imshow(mean)
```

```python
#2. Median Filter
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/Screenshot 2026-02-26 140552.png")

median = cv2.medianBlur(img, 3)

print("Original Image")
cv2_imshow(img)

print("Median Filter")
cv2_imshow(median)
```

```python
#3. Gaussian Filter
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/Screenshot 2026-02-26 140552.png")

gaussian = cv2.GaussianBlur(img, (3,3), 0)

print("Original Image")
cv2_imshow(img)

print("Gaussian Filter")
cv2_imshow(gaussian)
```

```python
#4. Spatial Filtering
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/Screenshot 2026-02-26 140552.png")

kernel = np.array([[1,1,1],
                   [1,1,1],
                   [1,1,1]], dtype=np.float32) / 9

filtered = cv2.filter2D(img, -1, kernel)

print("Original Image")
cv2_imshow(img)

print("Spatial Filter")
cv2_imshow(filtered)
```

```python
#5. Image Smoothing
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/Screenshot 2026-02-26 140552.png")

smooth = cv2.blur(img, (5,5))

print("Original Image")
cv2_imshow(img)

print("Smoothed Image")
cv2_imshow(smooth)
```

```python
#6. Image Sharpening
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/Screenshot 2026-02-26 140552.png")

kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

sharp = cv2.filter2D(img, -1, kernel)

print("Original Image")
cv2_imshow(img)

print("Sharpened Image")
cv2_imshow(sharp)
```
